```python
import pandas as pd
import numpy as np
```

```python
ttldf = pd.read_csv("ttldata.csv")
enadf = pd.read_csv("enadata.csv")
```

```python
ttldf['post'] = ttldf['heading'].str.cat(ttldf['post'], sep =" ")
ttldf = ttldf.drop(['heading'], axis=1)
enadf['post'] = enadf['heading'].str.cat(enadf['post'], sep =" ")
enadf = enadf.drop(['heading'], axis=1)
```

```python
ttldf = ttldf.sample(frac=0.1, replace=True, random_state=1)
enadf = enadf.sample(frac=0.3, replace=True, random_state=1)
df = pd.concat([ttldf,enadf],ignore_index = True)
```

```python
del df['Unnamed: 0']
df
```

```python
df['post'] = df[df['post'].str.len()>140]
df.dropna()
```
